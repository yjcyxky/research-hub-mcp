## Download pdfs from the pmc.

### Download all non-commercial pdf files from the pmc.

```python
from paper_parser.lib import PmcOaPdfDownloader, CacheManager

destdir = os.path.join(os.path.dirname(os.getcwd()), 'raw_data', 'pmc')

pmc_oa_pdf_downloader = PmcOaPdfDownloader(destdir=destdir, oa_non_comm=True)
metadata = pmc_oa_pdf_downloader.load_metadata()
files = list(metadata[0])

print("Number of files to download: {}".format(len(files)))

pmc_oa_pdf_downloader.download_files_parallel(files, n_jobs=40, skip=True, sleep=10)
```
