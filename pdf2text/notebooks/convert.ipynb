## Extract the related pdf file from the pmc directory

### Extract the related non-commercial pdf file from the pmc directory

```python
import shutil
from paper_parser.lib import PmcOaPdfDownloader, get_fdir, get_sdir

datadir = os.path.join(os.path.dirname(os.getcwd()), 'raw_data', 'pmc')
destdir = os.path.join(os.path.dirname(os.getcwd()), 'data')
pmc_oa_pdf_downloader = PmcOaPdfDownloader(destdir=datadir, oa_non_comm=True)
metadata = pmc_oa_pdf_downloader.load_metadata()

files = list(metadata[0])
pmids = [str(pmid).replace("PMID:", "") for pmid in list(metadata[3])]

for file, pmid in zip(files, pmids):
    if pmid == "nan":
        continue

    fdir = get_fdir(pmid)
    sdir = get_sdir(pmid)
    outputdir = os.path.join(destdir, fdir, sdir, pmid)
    outputfile = os.path.join(outputdir, "%s.pdf" % pmid)

    if not os.path.exists(outputdir):
        os.makedirs(outputdir)

    inputfile = os.path.join(datadir, file)
    print("Save %s into %s\n" % (file, outputdir))

    if os.path.exists(outputfile):
        print("%s exists, so skip it.\n" % outputfile)
        continue

    # Copy the file into the output directory
    shutil.copy(inputfile, outputfile)
```

### Extract the related commercial pdf file from the pmc directory

```python
import tarfile
import shutil
from paper_parser.lib import PmcOaPdfDownloader, get_fdir, get_sdir

datadir = os.path.join(os.path.dirname(os.getcwd()), 'raw_data', 'pmc')
destdir = os.path.join(os.path.dirname(os.getcwd()), 'data')
pmc_oa_pdf_downloader = PmcOaPdfDownloader(destdir=datadir, oa_non_comm=False)
metadata = pmc_oa_pdf_downloader.load_metadata()

files = list(metadata[0])
pmids = [str(pmid).replace("PMID:", "") for pmid in list(metadata[3])]

for file, pmid in zip(files, pmids):
    # TODO: Why don't have pmid?
    if pmid == "nan":
        continue

    fdir = get_fdir(pmid)
    sdir = get_sdir(pmid)
    outputdir = os.path.join(destdir, fdir, sdir, pmid)
    outputfile = os.path.join(outputdir, "%s.pdf" % pmid)

    if not os.path.exists(outputdir):
        os.makedirs(outputdir)

    inputfile = os.path.join(datadir, file)
    print("Save %s into %s" % (file, outputdir))

    if os.path.exists(outputfile):
        print("%s exists, so skip it.\n" % outputfile)
        continue

    # Extract a file from a tar.gz file
    with tarfile.open(inputfile, 'r:gz') as tar:
        files = tar.getnames()
        pdf_file = [file for file in files if file.endswith('.pdf')][0]
        print("Extract %s from %s." % (pdf_file, inputfile))
        tar.extract(pdf_file, outputdir)
        print("Move %s to %s.\n" % (pdf_file, outputfile))
        shutil.move(os.path.join(outputdir, pdf_file), outputfile)
        os.rmdir(os.path.join(outputdir, os.path.dirname(pdf_file)))
```
